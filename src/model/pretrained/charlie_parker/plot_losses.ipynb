```python
import torch
from matplotlib import pyplot as plt
```

```python
gen_losses = torch.load('generator_losses.pt')
gen_train_losses, gen_valid_losses = gen_losses['train_losses'], gen_losses['valid_losses']
dscr_losses = torch.load('discriminator_losses.pt')['losses']
```

```python
plt.rcParams['figure.figsize'] = [15, 5]

train_ax = plt.subplot(1, 2, 1)
train_ax.plot(gen_train_losses)
train_ax.set_title('Generator MLE Pretrain NLL')
train_ax.set_xlabel('Epochs')
train_ax.set_ylabel('Negative Log Likelihood (NLL)')

valid_ax = plt.subplot(1, 2, 2)
valid_ax.plot(gen_valid_losses)
valid_ax.set_title('Generator MLE Pretrain Validation NLL')
valid_ax.set_xlabel('Epochs')
valid_ax.set_ylabel('Negative Log Likelihood (NLL)')
```

```python
plt.rcParams['figure.figsize'] = [7, 5]

ax = plt.subplot(1, 1, 1)
ax.plot(dscr_losses)
ax.set_title('Discriminator Pretraining NLL')
ax.set_xlabel('Epochs')
ax.set_ylabel('Negative Log Likelihood (NLL)')
```
