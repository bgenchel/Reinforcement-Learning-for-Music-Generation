```python
import torch
from matplotlib import pyplot as plt
```

```python
data = torch.load('losses.pt')
gen_losses = [float(pt) for pt in data['adv_gen_losses']]
dscr_losses = [float(pt) for pt in data['adv_dscr_losses']]
```

```python
ax = plt.subplot(1, 1, 1)
ax.plot(gen_losses)
ax.set_title('Adversarial Generator Loss')
ax.set_xlabel('Epochs')
ax.set_ylabel('RL / Adversarial Loss')
```

```python
ax = plt.subplot(1, 1, 1)
ax.plot(dscr_losses)
ax.set_title('Adversarial Discriminator Loss')
ax.set_xlabel('Epochs')
ax.set_ylabel('Negative Log Likelihood (NLL)')
```
