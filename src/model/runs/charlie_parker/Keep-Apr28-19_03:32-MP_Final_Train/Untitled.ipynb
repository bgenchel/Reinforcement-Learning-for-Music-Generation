```python
import torch
import matplotlib.pyplot as plt

ld = torch.load('learning_data.pt', map_location="cpu")
print(ld.keys())
```

```python
plt.plot(ld['gen_updates'])
plt.title('Mean Policy Updates per Epoch')
plt.ylabel('Mean Update (-log(p(a|s)) * rewards)')
plt.xlabel('Adversarial Epoch')
```

```python
plt.plot(ld['rewards'])
plt.title('Mean Rewards per Epoch')
plt.xlabel('Adversarial Epochs')
plt.ylabel('Reward')
```

```python
plt.plot(ld['dscr_losses'])
plt.title('Mean Discriminator Loss')
plt.xlabel('Adversarial Epochs')
plt.ylabel('Negative Log Likelihood (NLL) Loss')
```
